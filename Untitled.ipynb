# Implementation of Multivariat Linear Regression using Gradient Descent

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

Generating some random data for the linear model to fit.

```python
true_parameters = np.array([3, 2, 5])
x = np.array([np.ones((100)), np.random.uniform(0.0, 10.0, (100)),np.random.uniform(0.0, 10.0, (100))])
y = true_parameters.dot(x) + np.random.uniform(-2.0, 2.0, (100))
data = np.array([*x,y])
data.shape
```

Initializing the two parameters to be estimated with random numbers.

```python
parameters = np.random.uniform(0.0, 1.0, (3))
```

```python
parameters
```

Comparing the data with the randomly initialized hypothesis.

```python
fig = plt.figure(figsize=(10, 10))
ax = Axes3D(fig)

predictions = parameters.dot(x)
ax.scatter(data[1], data[2], data[3], c="red")
ax.scatter(data[1], data[2], predictions)
plt.show()
```

Defining the loss function.

```python
def loss(data, parameters):
    predictions = parameters.dot(x)
    squared_errors = np.square(data[data.shape[0]-1] - predictions)
    summed_error = np.sum(squared_errors)
    normalized_error = summed_error / data.shape[1]
    return normalized_error
```

Defining the function for gradient calculation.

```python
def gradients(x, y, parameters):
    number_of_samples = x.shape[1]
    predictions = parameters.dot(x)
    error = (y - predictions)*(-1)
    gradients = x.dot(error)*(2/number_of_samples)
    return gradients
```

Running the gradient descent algorithm.

```python
lr = 0.01
epochs = 10

for i in range(epochs):
    parameters = parameters - lr * gradients(data[0:3], data[3], parameters)
    print(loss(data, parameters))
```

Visualizing the results, the predicted linear function now fits the data nicely!

```python
fig = plt.figure(figsize=(10, 10))
ax = Axes3D(fig)

predictions = parameters.dot(x)
ax.scatter(data[1], data[2], data[3], c="red")
ax.scatter(data[1], data[2], predictions)
plt.show()
```
